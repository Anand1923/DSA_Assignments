**3.Plot a bar graph of male vs female buyers participated in the sales**

**Pie Chart**

```python
c1 = pd.DataFrame()
c1['Count']=cars_data['Buyer Gender'].value_counts()
c1=c1.reset_index()
c1.rename(columns={'index':'Gender'},inplace=True)
c1
fig = plt.figure(figsize=[8,6])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Origin of the Cars')
#plotting pie chart
ax.pie(c1['Count'], labels=c1['Gender'],autopct='%0.1f%%', 
        shadow = True, startangle = 250
        )
fig.show()

```

**Bar Graph**

```python
plt.bar(c1['Gender'],c1['Count'])
plt.title('Male vs Female buyers participated in the sales',fontsize=13,fontweight='bold')
plt.xlabel('Gender',fontweight='bold')
plt.ylabel('Count',fontweight='bold')


```

**4.Sort Top 5 cars based on sales price**

```python
top_five=cars_data.sort_values( by='Sale Price',ascending=False).head(5)
top_five1=top_five[['Make','Model','Sale Price']]
top_five1
```

# #Extra

```python

plt.figure(figsize=(12,8))
x_axis=[]
x=np.arange(5)
w=0.4
avg=cars_data['Sale Price'].mean()
for i,j in zip(top_five1['Make'],top_five1['Model']):
  x_axis.append(i+' '+j+' AVG')
x_axis  
plt.bar(x-w/2,top_five1['Sale Price'],w,label='Top Five',color='Red')
plt.bar(x+w/2,avg,w,label='AVG',color='Blue')
plt.xticks(x,x_axis)
plt.legend()
```
